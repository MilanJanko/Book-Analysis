# Loading the book

```python
with open('miracle_in_the_andes.txt', 'r', encoding='utf-8') as file:
    book = file.read()
type(book)
```

## Check number of chapters with regex

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
find = re.findall(pattern, book)
len(find)
```

# Find sentences with word 'love'

```python
pattern = re.compile ('[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.')
find = re.findall(pattern, book)
find
```

# Most used words in the book

```python
pattern = re.compile('[a-zA-z]+')
find = re.findall(pattern, book.lower())
find[:8]
```

```python
word_dict = {}
for word in find:
    if word in word_dict.keys():
        word_dict[word] = word_dict[word] + 1
    else:
        word_dict[word] = 1

lista = [(value, key) for (key,value) in word_dict.items()]
lista = sorted(lista, reverse=True)
lista[:5]
```

## Most used words that are not articles

```python
import nltk
from nltk.corpus import stopwords

english_stopwords = stopwords.words('english')
```

```python
filtered = []
for count, word in lista:
    if word not in english_stopwords:
        filtered.append((word, count))
filtered
```

## Sentiment analyse: What is most positive/negative chapter?

```python
from nltk.sentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

# Chapter sentiment analyse

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]
```

```python
for number, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(number+1, scores)
```
